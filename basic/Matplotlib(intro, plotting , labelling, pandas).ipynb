#matplotlib  : Labeling

```python
jobcounts=df.job_title_short.value_counts()
plt.bar(jobcounts.index,jobcounts.values)                    # you can also go to matplotlib.org/cheatsheets
plt.title('postings by job title')
plt.ylabel('number of postings')
plt.xticks(rotation=45,ha='right')                             #rotate  x-axis labels by 45 degree, ha is horizontal alignment of labels
plt.xlabel('job title')
plt.show()
```

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
jobSalary=df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
jobSalary.plot(kind='barh')
plt.xlabel('salary ($usd)')
plt.ylabel('')
plt.title('median salary by job title')
plt.show()                                         # to run plot method it can be only called on a series
```

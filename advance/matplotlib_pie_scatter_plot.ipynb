```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast
dataset=load_dataset('lukebarousse/data_jobs')

df=dataset['train'].to_pandas()

df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.head()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie')
plt.title('work from home')
plt.ylabel('')
```

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title('job titles')
plt.ylabel('')
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',startangle=90,autopct='%1f')
plt.title('work from home')
plt.ylabel('')
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f')
plt.title('work from home')
plt.ylabel('')
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f%%')
plt.title('work from home')
plt.ylabel('')
```

```python
fig,ax=plt.subplots(1,3)

dict_col={
 'job_work_from_home':'Work from Home',                    #column:title
 'job_no_degree_mention':'Job Degree Req.',
 'job_health_insurance':'Health Insurance Offered'
}

for i ,(column,title) in enumerate(dict_col.items()):                           #if we run enumerate func it provide the index back
 # print(i,column,title)
  ax[i].pie(df[column].value_counts(),startangle=90,autopct='%1.1f%%',labels=['False','True'])
  ax[i].set_title(title)



plt.show()
```

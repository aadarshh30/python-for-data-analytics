```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset=load_dataset('lukebarousse/data_jobs')

df=dataset['train'].to_pandas()

df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_US=df[df['job_country']=='United States'].copy()
```

```python
df_US['job_posted_month']=df_US['job_posted_date'].dt.strftime('%B')           #gives month like june , july etc

df_US
```

```python
df_US_pivot=df_US.pivot_table(columns='job_title_short', index='job_posted_month',aggfunc=('size'))
```

```python
df_US_pivot.reset_index(inplace=True)
```

```python
df_US_pivot['month_no']=pd.to_datetime(df_US_pivot['job_posted_month'],format='%B').dt.month
```

```python
df_US_pivot.sort_values(by='month_no', inplace=True)
```

```python
df_US_pivot.set_index('job_posted_month', inplace=True)
```

```python
df_US_pivot.drop(columns='month_no', inplace=True)
```

```python
df_US_pivot
```

```python
df_US_software_pivot=pd.read_csv('https://lukeb.co/software_csv',index_col='job_posted_month')           #data publicly available
```

```python
df_US_software_pivot
```

```python
df_US_merged=df_US_pivot.merge(df_US_software_pivot,on='job_posted_month')
df_US_merged
```

```python
top_5=df_US_merged.sum()
top_5=top_5.sort_values(ascending=False).head().index
top_5=top_5.to_list()
```

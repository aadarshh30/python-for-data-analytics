```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset=load_dataset('lukebarousse/data_jobs')

df=dataset['train'].to_pandas()

df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_US=df[df['job_country']=='United States'].copy()
```

```python
df_US
```

```python
df_US['job_posted_month']=df_US['job_posted_date'].dt.strftime('%B')           #gives month like june , july etc

df_US
```

```python
df_US_pivot=df_US.pivot_table(columns='job_title_short', index='job_posted_month',aggfunc=('size'))



# df_US.pivot_table(index='job_title_short', columns='job_posted_month',aggfunc=('size'))
```

```python
df_US_pivot.reset_index(inplace=True)
```

```python
df_US_pivot['month_no']=pd.to_datetime(df_US_pivot['job_posted_month'],format='%B').dt.month
```

```python
df_US_pivot.sort_values(by='month_no', inplace=True)
```

```python
df_US_pivot.set_index('job_posted_month', inplace=True)
```

```python
df_US_pivot.drop(columns='month_no', inplace=True)
```

```python
df_US_pivot
```

```python
df_US_pivot.plot(kind='line')
```

```python
top_3=df_US['job_title_short'].value_counts().head(3)

top_3=top_3.index.tolist()
```

```python
df_US_pivot[top_3].plot(kind='line')
```

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('monthly job posting for top three jobs in us')
plt.xlabel('2023')
plt.ylabel('job counts')
plt.legend()
plt.show()
```
